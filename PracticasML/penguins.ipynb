```python
import pandas as pd

nomb_fich='C:/Users/migue/Downloads/penguins_mod.csv'
penguins = pd.read_csv(nomb_fich)

print(penguins.columns.values)
```

3. Emplearemos la función lmplot() de seaborn y la función grid() de pyplot, por lo que importaremos las librerías y módulos correspondientes. El código requerido es:

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lmplot('bill_length_mm','bill_depth_mm',data=penguins,hue='species',fit_reg=False)
plt.grid()
```
